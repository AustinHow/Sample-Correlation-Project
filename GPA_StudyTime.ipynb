```python
df = pd.read_csv(r"C:\Users\Austin\Desktop\gpa_study_hours.csv")
print(df)
```

```python
df=((df-df.min())/(df.max()-df.min()))*20
```

```python
sns.regplot(x = "study_hours", y = "gpa", data = df )
```
